```python
# read first 5 rows of speed flow data
# filename = "/Users/vree/Documents/UTwente/Masters UTwente/Data Science/DM Project/Speed-flow/Maart/intensiteit-snelheid-rotterdam-15-21-Maart/intensiteit-snelheid-rotterdam-15-21-Maart.csv"
filename = "/Users/vree/Documents/UTwente/Masters UTwente/Data Science/DM Project/Speed-flow/Maart/intensiteit-snelheid-rotterdam-22-29-Maart/intensiteit-snelheid-rotterdam-22-29-Maart.csv"
speed_flow_example = pd.read_csv(filename, nrows=5)
speed_flow_example
```

```python

# Create an empty DataFrame with the same columns as the original data
unique_id_speed_flow = pd.DataFrame(columns=speed_flow_example.columns)

# Set to track seen 'meetlocatie_id'
seen_ids = set()

# Read the file in chunks
for chunk in pd.read_csv(filename, chunksize=10000):
    # Filter the chunk to only include new, unique 'meetlocatie_id'
    new_rows = chunk[~chunk['id_meetlocatie'].isin(seen_ids)]
    
    # Update the set of seen ids
    seen_ids.update(new_rows['id_meetlocatie'])
    
    # Append new unique rows
    unique_id_speed_flow = pd.concat([unique_id_speed_flow, new_rows.drop_duplicates(subset=['id_meetlocatie'])])

# Display the shape of the final DataFrame with unique 'meetlocatie_id'
print(unique_id_speed_flow.shape)
```

```python
# now go to the incidents dataframe, and check whether the cameraid in the json rows for columns: before, after
# is in the unique_id_speed_flow dataframe, as id_meetlocatie
# add the number of cameras before and after the incident to the incidents dataframe as a column
import json

for i in incidents_df.index:  # Iterate over the DataFrame index instead of a range
    # Parse JSON data from 'before' and 'after' columns
    json_data_before = json.loads(incidents_df.at[i, 'before']) if pd.notna(incidents_df.at[i, 'before']) else {}
    json_data_after = json.loads(incidents_df.at[i, 'after']) if pd.notna(incidents_df.at[i, 'after']) else {}
    
    
    
    incidents_df.at[i, 'n_cameras_before'] = len(json_data_before)
    incidents_df.at[i, 'n_cameras_after'] = len(json_data_after)
    
    

    # Extract camera ids from JSON keys, handling the case where JSON data might be empty
    camera_ids_before = list(json_data_before.keys()) if json_data_before else []
    camera_ids_after = list(json_data_after.keys()) if json_data_after else []

    # Count the number of cameras before and after the incident that are in 'unique_id_speed_flow'
    num_cameras_before = unique_id_speed_flow[unique_id_speed_flow['id_meetlocatie'].isin(camera_ids_before)].shape[0]
    num_cameras_after = unique_id_speed_flow[unique_id_speed_flow['id_meetlocatie'].isin(camera_ids_after)].shape[0]

    # Add the number of cameras before and after the incident to the incidents dataframe
    incidents_df.at[i, 'num_cameras_bfre_in_speedflow'] = num_cameras_before
    incidents_df.at[i, 'num_cameras_after_in_speedflow'] = num_cameras_after

# Display or return the updated DataFrame
incidents_df

    
```

```python
# see whether there's camera's missing, which can be done by n_cameras_before-n_cameras_bfre_in_speedflow and n_cameras_after-num_cameras_after_after,
# add up the number and present as 'missing_cameras' and add as column
incidents_df['missing_cameras'] = incidents_df['n_cameras_before'] - incidents_df['num_cameras_bfre_in_speedflow'] + incidents_df['n_cameras_after'] - incidents_df['num_cameras_after_in_speedflow']
incidents_df

```
