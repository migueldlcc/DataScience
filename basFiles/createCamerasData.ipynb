```python
file_name = '/Users/vree/Documents/UTwente/Masters UTwente/Data Science/DM Project/Speed-flow/Maart/intensiteit-snelheid-rotterdam-1-7-Maart/intensiteit-snelheid-rotterdam-1-7-Maart.csv'

cameraString = ''
with open('camerasForTest.txt') as f:
    cameraString = f.readlines()[0]
    cameraString = cameraString.replace('\n', '' )
    cameraString = cameraString.split(',')
    
    
print(cameraString)


camera_ids = cameraString

camera_data = pd.DataFrame()

for chunk in pd.read_csv(file_name, chunksize=10000):
    chunk = chunk[chunk['id_meetlocatie'].isin(camera_ids)]
    camera_data = pd.concat([camera_data, chunk])
    

camera_data.to_csv('camerasData2.csv', index=False)

camera_data.head()




```

```python


print('cmaera data shape: ', camera_data.shape)

# get mean date for start_meetperiode
camera_data['start_meetperiode'] = pd.to_datetime(camera_data['start_meetperiode'])
# drop not available
camera_data.head()

camera_data = camera_data.dropna(subset=['start_meetperiode'])
print('mean date: ', camera_data['start_meetperiode'].mean())

# get the type of the coluns
print('data types: ', camera_data.dtypes)

# filter out mean time -2.5h and mean time +2.5h
# 5 hours
start_time = camera_data['start_meetperiode'].mean() - pd.Timedelta('10 hours')
print('start time: ', start_time)

end_time = camera_data['start_meetperiode'].mean() + pd.Timedelta('10 hours')
print('end time: ', end_time)
camera_data = camera_data[(camera_data['start_meetperiode'] > start_time) & (camera_data['start_meetperiode'] < end_time)]


print('shape after data filter: ', camera_data.shape)
camera_data.to_csv('cameradata2.csv', index=False)
```
