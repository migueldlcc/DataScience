```python
import pandas as pd
import folium
```

<h1>Bridges Table</h1>

```python
df4 = pd.read_csv('C:/Users/migue/source/repos/DataScience/datasets/status/sb-bruggen-export_rotterdam 1 maart - 5 april.zip', compression='zip')
df5 = pd.read_csv('C:/Users/migue/source/repos/DataScience/datasets/status/sb-bruggen-export_rotterdam 27 april - 31 mei.zip', compression='zip')
print(df4.columns)
print(df5.columns)
```

```python
bridge_march_df = df4[['id', 'naam', 'aangekondigd', 'geopend', 'gesloten', 'lengtegraad', 'breedtegraad']]
bridge_march_df = bridge_march_df.drop_duplicates(ignore_index=True,keep='last')
bridge_may_df = df5[['id', 'naam', 'aangekondigd', 'geopend', 'gesloten', 'lengtegraad', 'breedtegraad']]
bridge_may_df = bridge_may_df.drop_duplicates(ignore_index=True,keep='last')

# Convert 'naam' column to string
bridge_march_df['naam'] = bridge_march_df['naam'].astype(str)
bridge_may_df['naam'] = bridge_may_df['naam'].astype(str)

# Filter rows where 'naam' column contains 'Van Brienenoordbrug'
van_brienenoordbrug_march_df = bridge_march_df[bridge_march_df['naam'].str.contains("Van Brienenoordbrug", na=False, case=False)]
van_brienenoordbrug_may_df = bridge_may_df[bridge_may_df['naam'].str.contains("Van Brienenoordbrug", na=False, case=False)]
```

```python
# Concatenate the datasets
bridges_df = pd.concat([van_brienenoordbrug_march_df, van_brienenoordbrug_may_df], ignore_index=True)
bridges_df = bridges_df.drop_duplicates(ignore_index=True,keep='last')
bridges_df = bridges_df.drop('naam', axis=1)

# Convert 'aangekondigd' to datetime format
bridges_df['aangekondigd'] = pd.to_datetime(bridges_df['aangekondigd'])

bridges_df = bridges_df.sort_values(by='aangekondigd')
bridges_df.reset_index(drop=True, inplace=True)
                                            
print(len(bridges_df))
bridges_df.head()


```

```python
# Create a map centered around an average location or a specific bridge action location
map_center_latitude = bridges_df['breedtegraad'].mean()
map_center_longitude = bridges_df['lengtegraad'].mean()
m = folium.Map(location=[map_center_latitude, map_center_longitude], zoom_start=12)

# Add markers to the map for each bridge action
for index, row in bridges_df.iterrows():
    folium.Marker(
        [row['breedtegraad'], row['lengtegraad']],
        popup=f"ID: {row['id']}\nAangekondigd: {row['aangekondigd']}\nGeopend: {row['geopend']}\nGesloten: {row['gesloten']}",
        tooltip=f"Bridge ID: {row['id']}"
    ).add_to(m)

# Display the map
m
```

<h1>Work Roads</h1>

```python
df6 = pd.read_csv('C:/Users/migue/source/repos/DataScience/datasets/status/sb-wegwerkzaamheden-export _ Rotterdam 1 maart - 31 augustus.zip', compression='zip')
print(df6.columns)
```

```python
work_roads_df = df6[['id', 'type', 'starttijd', 'eindttijd', 'locatie']]
print(len(work_roads_df))
work_roads_df.head()
```

```python
# Ensure locatie is treated as a string for regex operations
work_roads_df['locatie'] = work_roads_df['locatie'].astype(str)

# Extract longitude and latitude from the 'locatie' column
work_roads_df[['longitude', 'latitude']] = work_roads_df['locatie'].str.extract(r'POINT\(\s*(\d+\.\d+)\s+(\d+\.\d+)\s*\)')
work_roads_df = work_roads_df.drop('locatie', axis=1)
print(len(work_roads_df))
unique_id_count = work_roads_df['id'].nunique()
print(unique_id_count)

# Drop rows with NaN values in 'latitude' or 'longitude'
work_roads_df.dropna(subset=['latitude', 'longitude'], inplace=True)

print(len(work_roads_df))
work_roads_df = work_roads_df.drop_duplicates(subset=['id', 'starttijd', 'eindttijd'], ignore_index=True,keep='last')
print(len(work_roads_df))

# Convert 'starttijd' to datetime format
work_roads_df['starttijd'] = pd.to_datetime(work_roads_df['starttijd'])

# Now sort the DataFrame by 'starttijd'
work_roads_df = work_roads_df.sort_values(by='starttijd')

# Reset the index if you want to get rid of the old index
work_roads_df.reset_index(drop=True, inplace=True)

# Display the DataFrame with the new columns
work_roads_df.head()
```
